```python
import requests
from bs4 import BeautifulSoup
```

```python
urls = ['https://www.amazon.in/gp/new-releases/luggage/ref=zg_bs_pg_1?ie=UTF8&pg=1', 'https://www.amazon.in/gp/new-releases/luggage/ref=zg_bs_pg_2?ie=UTF8&pg=2', 'https://www.amazon.in/gp/new-releases/luggage/ref=zg_bs_pg_3?ie=UTF8&pg=3','https://www.amazon.in/gp/new-releases/luggage/ref=zg_bs_pg_4?ie=UTF8&pg=4','https://www.amazon.in/gp/new-releases/luggage/ref=zg_bs_pg_5?ie=UTF8&pg=5']
#'https://www.amazon.in/gp/bestsellers/books/ref=zg_bs_pg_'+str(2)+'?ie=UTF8&pg=%27'
HEADERS = ({"User-Agent":"Mozilla/5.0 (Windows NT 10.0; Win64; x64; rv:66.0) Gecko/20100101 Firefox/66.0", "Accept-Encoding":"gzip, deflate", "Accept":"text/html,application/xhtml+xml,application/xml;q=0.9,*/*;q=0.8", "DNT":"1","Connection":"close", "Upgrade-Insecure-Requests":"1"})

soups = []
for url in urls:
  r = requests.get(url,headers=HEADERS)
  content = r.content
  soups.append(BeautifulSoup(content, 'lxml'))
```

```python
for soup in soups:
  print(soup.title)
```

```python
for soup in soups:
  print(soup.body)
```

```python
price = []
for soup in soups:
  for i in soup.findAll('span',attrs={'class':"a-size-base a-color-price"}):
    price.append(i.text)
price
```

```python
for soup in soups:
  print(soup.select('span'))
```

```python
name = []
for soup in soups:
  for i in soup.findAll('span'):
    for j in i.findAll('div',attrs={'class':["_cDEzb_p13n-sc-css-line-clamp-3_g3dy1",'_cDEzb_p13n-sc-css-line-clamp-4_2q2cc','_cDEzb_p13n-sc-css-line-clamp-5_2l-dX']}):
      if j == None:
        name.append("nan")
      else:
        name.append(j.text)
len(name)
```

```python
review = []
for soup in soups:
  for i in soup.findAll('span',attrs={'class':"a-icon-alt"}):
    if i not in soup.findAll('span',attrs={'class':"a-icon-alt"}):
        review.append("nan")
    else:
        review.append(i.text)
review
```

```python
for soup in soups:
  for i in soup.findAll('span'):
    print(i)
```

```python
likes = []
for soup in soups:
  for i in soup.findAll('span',attrs={'class':"a-size-small"}):
    likes.append(i.text)
len(likes)
```

```python
price_on_size_range = []
for soup in soups:
  for i in soup.findAll('span',attrs={'class':"a-size-base a-color-price"}):
    price_on_size_range.append(i.text)
len(price_on_size_range)
```

```python
import pandas as pd
```

```python
df = pd.DataFrame({"Name_of_Product":pd.Series(name),"Price":pd.Series(price_on_size_range),"Liked_by_customers":pd.Series(likes),'Reviews':pd.Series(review)})
```

```python
df.dropna(inplace=True)
```
